## Representacion gràfica de datos.

Si a la funcion de trasado se le da una matriz de datos, la usarà como coordenadas en el **eje vertical**, y utlizarà el indice de cada punto de datos en el array como la coordenada horizontal.

```python
plt.plot([1, 2, 5, 7, 8, 3, 1])
plt.show()
```

Tambien se puede proporcinar dos matrices: una para el eje horzontal y otra para el eje vertical.

```python
plt.plot([-3, -1, 0, 4, 7], [1, 4, 6, 7, 8])
plt.show()
```

Pueden modificarse las longitudes de los ejes para que la figura no se vea tan ajustada.

```python
plt.plot([-3, -1, 0, 4, 7], [1, 4, 6, 7, 8])
plt.axis([-4 , 8, 0, 10]) #[xmin, xmax, ymin, ymax]
plt.show()
```

Se sigue el mismo prodedimiento para pintar una funciona metematica.

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
plt.plot(x, y)
plt.show()
```

 Tambien puede modificarse el estilo de la gràfica para que cotenga mas infomracion.

```python
plt.plot(x, y)
plt.title("Square Function")
plt.xlabel("Chido")
plt.ylabel("y = **2")
plt.grid(True)
plt.show()
```

Pueden superponerse graficas y cambir el estilo de las funciones.

```python

import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1
plt.title("Square Function")
plt.xlabel("Chido")
plt.ylabel("y = **2")
plt.grid(True)
plt.plot(x, y, 'b--', x, y2, 'g')
plt.show()
```

```python
# Seprarando en diferentes lienas las funciones.
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1
plt.title("Square Function")
plt.xlabel("Chido")
plt.ylabel("y = **2")
plt.grid(True)
plt.plot(x, y, 'b--')
plt.plot(x, y2, 'g')#EL b-- es para colocar un color azul y qeu la linea sea punteada.
plt.show()
```

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1
plt.title("Square Function")
plt.xlabel("Chido")
plt.ylabel("y = **2")
plt.grid(True)
plt.plot(x, y, 'b--', label = "x**2") # la etioqueta (label) le indica al usuario que es cada cosa.
plt.plot(x, y2, 'g', label = "x + 1")
plt.legend(loc = "best") # La situa en la mejor localizacion.
plt.show()
```

Tambien puede crearse os graficas que no se superpongand, estas graficas se organizan en un **grid** y se denominan **subplots**.

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.subplot(1, 2, 1) # 1 Rows, 2 columns, 1st subplot
plt.plot(x, y, 'b--')

plt.subplot(1, 2, 2) # 1 Rows, 2 columns, 2nd subplots
plt.plot(x, y2, 'g')

plt.show()
```

Para que las graficas no queden tan ajsutadas, se puede hacer la figura mas grande.

```python
plt.figure(figsize = (14, 6))

plt.subplot(1, 2, 1) # 1 Rows, 2 columns, 1st subplot
plt.plot(x, y, 'b--')
plt.title("Square Function", fontsize = 14)
plt.xlabel("Chido", fontsize = 14)
plt.ylabel("y = **2", fontsize = 14)
plt.grid(True)


plt.subplot(1, 2, 2) # 1 Rows, 2 columns, 2nd subplots
plt.plot(x, y2, 'g')
plt.title("Lineal Function", fontsize = 14)
plt.xlabel("Chido 2", fontsize = 14)
plt.ylabel("y2 = **2", fontsize = 14)
plt.grid(True)


plt.show()
```

## Scatter (dispersion) Plots

```python
from numpy.random import rand
x, y = rand(2, 100)
plt.scatter(x, y)
plt.show()
```

```python
from numpy.random import rand
x, y = rand(2, 100)
x2, y2 = rand(2, 100)
plt.scatter(x, y, c = "red")
plt.scatter(x2, y2, c = "blue")
plt.show()
```

## Histogramas

```python
data = [1, 1.1, 1.8, 2, 2.1, 3.2, 3, 3, 3, 3]
plt.subplot(211)
plt.hist(data, bins = 10, rwidth = 0.8)

plt.subplot(212)
plt.hist(data, bins = [1, 1.5, 2, 2.5, 3], rwidth = 0.95)

plt.xlabel('Value')
plt.ylabel('Frequency')

plt.show()
```
